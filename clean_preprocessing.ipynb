Phân tích kiểu dữ liệu từng cột trong data

```python
#in ra kiểu dữ liệu của các cột
print("Kiểu dữ liệu của các cột trong DataFrame: ")
print(df.dtypes)
```

#Ép kiểu dữ liệu

```python
#Ép kiểu dữ liệu cho các cột
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['seniority'] = pd.to_numeric(df['seniority'], errors='coerce').astype('float32')


```

Tìm kiếm các cột có đữ liệu bị thiếu, loại bỏ các dòng có trên 2 giá trị bị thiếu

```python
# Đếm số dòng vi phạm (có trên 2 giá trị thiếu)
so_dong_bi_xoa = df[df.isnull().sum(axis=1) < 2].shape[0]

print(f"Tổng số dòng hiện tại: {df.shape[0]}")
print(f"Số dòng sẽ bị xóa: {so_dong_bi_xoa}")
print(f"Tỷ lệ mất dữ liệu: {(so_dong_bi_xoa / df.shape[0]) * 100:.2f}%")# Giá trị trên 5% thì cần xem xét lại

    
```

Xóa toàn bộ các hàng có 1 giá trị bị trống

```python

df = df[df.isnull().sum(axis=1) != 1]
print(f"Số dòng sau khi loại bỏ: {df.shape[0]}")
```

Vẽ biểu đồ đường tìm giá trị ngoại lai trong tuổi của người giao dịch

```python
#Vẽ biểu đồ đường tìm giá trị ngoại lai trong tuổi của người giao dịch
plt.plot(df['age'])
plt.title('Biểu đồ đường của tuổi người giao dịch')
plt.xlabel('Chỉ số dòng')
plt.ylabel('Tuổi')
plt.show()
```

```python
#Xem xét các giá trị ngoại lai trong cột 'age'
outliers = df[(df['age'] < 18) | (df['age'] > 100)]
print("Các giá trị ngoại lai trong cột 'age':")
print(outliers)
```

```python
#Xóa hết các dòng có giá trị ngoại lai >100
df = df[df['age'] <= 100].dropna(subset=['age'])
print(f"Số dòng sau khi loại bỏ giá trị ngoại lai: {df.shape[0]}")
```

Xóa hết các dòng có giá trị ngoại lai age <18

```python
df = df[df['age'] >= 18].dropna(subset=['age'])
print(f"Số dòng sau khi loại bỏ giá trị ngoại lai: {df.shape[0]}")
```

```python
#in ra 5 giá trị seniority lớn nhất trong dataframe
#Seniority tính theo tháng, age tính theo năm
#seniority âm không được xuất hiện trong dữ liệu
print("20 giá trị seniority và age đầu tiên trong DataFrame:")
print(df[['seniority', 'age']].sort_values(by='age', ascending=False).head(30))



```

```python
#xóa giá trị seniority âm
#Tìm index của các dòng có seniority âm hoặc bị rỗng (NaN)
#Lưu ý: Chỉ lấy index thôi nên tốn rất ít RAM
indices_to_drop = df[ (df['seniority'] < 0) | (df['seniority'].isna()) ].index

#Xóa các dòng này trực tiếp trên DataFrame gốc
df.drop(indices_to_drop, inplace=True)
print('Đã xóa thành công các dòng có giá trị seniority âm hoặc bị rỗng.')
# Giải phóng bộ nhớ không sử dụng
```

```python
def check_seniority_over_age(df):
    # SỬA 1: Dùng phép so sánh (>) để tạo điều kiện lọc
    # Tìm những người có thâm niên (tháng) lớn hơn số tuổi quy đổi ra tháng
    condition = df['seniority'] > (df['age'] * 12)
    
    seniors_over_age = df[condition]

    # SỬA 2: Dùng .empty để kiểm tra có dữ liệu hay không
    if not seniors_over_age.empty:
        print(f"Phát hiện {len(seniors_over_age)} dòng có thâm niên lớn hơn tuổi đời:")
        # Chỉ in 5 dòng đầu tiên để xem mẫu cho đỡ rối
        print(seniors_over_age[['seniority', 'age']].head())
    else:
        print("Dữ liệu hợp lý: Không có dòng nào thâm niên lớn hơn tuổi.")

# SỬA 3: Chỉ cần gọi hàm, không cần print bọc ngoài
check_seniority_over_age(df)
```

Tìm kiếm top 10 giá trị bị thiếu trong dataframe và vẽ column

```python
#Tìm kiếm top 10 giá trị bị thiếu trong dataframe và vẽ bar chart
missing_values = df.isnull().sum().sort_values(ascending=False).head(10)
print("Top 10 giá trị bị thiếu trong DataFrame:")
print(missing_values)
missing_values.plot(kind='bar')
plt.title('Top 10 giá trị bị thiếu trong DataFrame')
plt.xlabel('Cột')
plt.ylabel('Số giá trị bị thiếu')
plt.show()
```

```python
#Với các cột quá nhiều dữ liệu trống như 'spouse_employee', 'ult_fec_cli_It',  loại bỏ hoàn toàn cột đó 
cols_to_drop = ['spouse_employee', 'ult_fec_cli_1t']
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
print("Đã xóa xong các cột rác. Đang giải phóng RAM...")
#Với cột 'income' có giá trị bị thiếu, ta sẽ thay thế bằng giá trị median của cột đó
median_income = df['income'].median()
df['income'].fillna(median_income, inplace=True)
print("Đã thay thế giá trị bị thiếu trong cột 'income' bằng median.")
```

```python
#Xử lý các cột segment, entry_channel xoa het cac dong bi thieu gia tri o cac cot nay
cols_to_clean = ['segment', 'entry_channel']
df.dropna(subset=cols_to_clean, inplace=True)
print("Đã xóa xong các dòng bị thiếu giá trị trong cột 'segment' và 'entry_channel'.")
#Voi cac col con lai, thay the gia tri bi thieu bang mode cua cot do
cols_to_fill_mode = ['country_of_residence', 'relation_index_1month', 'province_code', 'type_of_relation_1month']
for col in cols_to_fill_mode:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)
    print(f"Đã thay thế giá trị bị thiếu trong cột '{col}' bằng mode.")

```

XUat file sau khi lam sach

```python
export_filename = 'cleaned_data.csv'
df.to_csv(export_filename, index=False)
print(f"Dữ liệu đã được làm sạch và lưu vào tệp '{export_filename}'.")
```
